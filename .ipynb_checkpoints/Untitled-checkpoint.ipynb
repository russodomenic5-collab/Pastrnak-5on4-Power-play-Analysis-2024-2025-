```python
# --- Step 1.1: Initial Setup and Data Load ---

# I am importing the necessary libraries for data handling (Pandas) and numerical operations (NumPy).
import pandas as pd
import numpy as np

# --- Define Constants ---
# I am defining constants for player ID, player name, and team abbreviation.
# These constants make the code cleaner and easier to filter later.
PASTRNAK_NAME = 'David Pastrnak'
BRUINS_TEAM = 'BOS' 
PASTRNAK_ID = 8477956 

# --- Load DataFrames ---

# I am loading the Skaters Data (aggregated rates and TOI).
# This file is used to calculate the Per 60 rates for the bar chart.
df_skaters = pd.read_csv("skaters_20242025.csv")

# I am loading the Raw Shot Data (coordinates and individual events).
# This file is used to create the Shot Map visualization.
df_raw_shots = pd.read_csv("shots_20242025.csv")

# I am loading in the Team Level Data for all 32 NHL teams for the 2024-2025 season
# This will be used to find the Bruins 2024-2025 team total 5v4 ice time, which we will subtract Pastrnaks ice time from to find the true OffIce icetime.
df_team = pd.read_csv("team_20242025.csv")
df_team['icetime'] = pd.to_numeric(df_team['icetime'], errors='coerce')

# I am initializing the final summary table variable.
final_summary_df = pd.DataFrame() 

# I am performing a quick audit to confirm the files loaded successfully.
print("Step 1.1: Data Loading Complete.")
print(f"Skaters Data Rows: {len(df_skaters)}")
print(f"Raw Shot Data Rows: {len(df_raw_shots)}")
print(f"Team Level Data Rows: {len(df_team)}")
```

```python
# --- Step 1.2: Filter Data and Isolate Pastrnak's Rates ---

# --- 1. Filter Data for Relevance (SQL WHERE equivalent) ---

# I am defining the filtering step, which is the Python equivalent of a SQL 'WHERE' clause.
# I am isolating all players who played for the Bruins in the 5-on-4 power play situation.
df_pp = df_skaters[
    (df_skaters['team'] == BRUINS_TEAM) & 
    (df_skaters['situation'] == '5on4') 
].copy() 

# --- 2. Calculate Pastrnak's ON-ICE Rates ---

# I am isolating David Pastrnak's data row.
df_pastrnak = df_pp[df_pp['name'] == PASTRNAK_NAME].copy()

# Total time Pastrnak played on the power play, converted from seconds to minutes.
pastrnak_toi_min = df_pastrnak['icetime'].iloc[0] / 60 

# I am calculating the Per 60 minute rates for Pastrnak's unit. 
# MoneyPuck column names are confirmed: OnIce_F_shotAttempts (Corsi), OnIce_F_xGoals (xG), OnIce_F_goals (GF), I_F_goals (Individual Goals).
pastrnak_cf60 = (df_pastrnak['OnIce_F_shotAttempts'].iloc[0] / pastrnak_toi_min) * 60
pastrnak_xg60 = (df_pastrnak['OnIce_F_xGoals'].iloc[0] / pastrnak_toi_min) * 60
pastrnak_gf60 = (df_pastrnak['OnIce_F_goals'].iloc[0] / pastrnak_toi_min) * 60
pastrnak_g60 = (df_pastrnak['I_F_goals'].iloc[0] / pastrnak_toi_min) * 60 

print("Step 1.2: Pastrnak's ON-ICE Rates Successfully Isolated.")
print(f"True ON CF/60: {pastrnak_cf60:.2f}")
print(f"True ON xG/60: {pastrnak_xg60:.2f}")
print(f"True ON GF/60: {pastrnak_gf60:.2f}")
print(f"TRUE ON G/60: {pastrnak_g60:.2f}")
```

```python
# --- Step 1.3: CORRECTED Calculate Team Residual OFF-ICE Rates (All 3 Metrics) ---

# 1. Get the Bruins' total 5v4 TOI (in minutes) and Goals For (GF) from the team data.
bruins_5v4_row = df_team[(df_team['team'] == BRUINS_TEAM) & (df_team['situation'] == '5on4')]

# Get TOTAL Team TOI (using 'iceTime')
team_5v4_toi_minutes = bruins_5v4_row['icetime'].iloc[0] / 60

# Get TOTAL Team Goals For (using 'goalsFor' as confirmed by user)
team_5v4_total_gf = bruins_5v4_row['goalsFor'].iloc[0] # <--- UPDATED COLUMN NAME

# 2. Calculate the TRUE OFF-ICE TOI (Team Total - Pastrnak's On-Ice)
true_off_ice_toi_min = team_5v4_toi_minutes - pastrnak_toi_min

# 3. Calculate the Total Off-Ice Events (CF, xG, and now GF)

# CF & xG are pulled from the existing 'OffIce_F_' columns in Pastrnak's row
off_ice_total_cf = df_pastrnak['OffIce_F_shotAttempts'].iloc[0]
off_ice_total_xg = df_pastrnak['OffIce_F_xGoals'].iloc[0]

# GF is calculated using the residual method (Team Total Goals - Pastrnak's OnIce Goals)
pastrnak_on_ice_gf = df_pastrnak['OnIce_F_goals'].iloc[0]
off_ice_total_gf = team_5v4_total_gf - pastrnak_on_ice_gf # <--- RESIDUAL CALCULATION

# 4. Calculate the Per 60 rates for the TRUE Residual OFF-ICE unit.
team_avg_cf60 = (off_ice_total_cf / true_off_ice_toi_min) * 60
team_avg_xg60 = (off_ice_total_xg / true_off_ice_toi_min) * 60
team_avg_gf60 = (off_ice_total_gf / true_off_ice_toi_min) * 60 # <--- NEW GF/60 CALCULATION

print("\nStep 1.3: TRUE Residual OFF-ICE Rates Calculated (CF/60, xG/60, and GF/60).")
print(f"Total Team 5v4 Goals For: {team_5v4_total_gf}")
print(f"Pastrnak On-Ice Goals For: {pastrnak_on_ice_gf}")
print(f"True Residual OFF-ICE Goals For: {off_ice_total_gf} (Calculated)")
print("---")
print(f"True OFF CF/60: {team_avg_cf60:.2f}")
print(f"True OFF xG/60: {team_avg_xg60:.2f}")
print(f"True OFF GF/60: {team_avg_gf60:.2f}")
```

```python
# --- Step 1.4: Consolidate into Final Summary Table ---

# I am structuring the data for my final bar chart visualization.
summary_data = {
    'Unit': ['Pastrnak On', 'Pastrnak Off'],
    'CF/60_Rate': [pastrnak_cf60, team_avg_cf60],
    'xG/60_Rate': [pastrnak_xg60, team_avg_xg60],
    'GF/60_Rate': [pastrnak_gf60, team_avg_gf60],
    'G/60_Rate': [pastrnak_g60, 0.0] 
}
# I am creating the final DataFrame that holds the results for the bar chart.
final_summary_df = pd.DataFrame(summary_data)

# I want to confirm the structure of my final results table.
print("\nStep 1.4: Final Consolidated Summary Table.")
print(final_summary_df)
```

```python
# --- Step 1.5: Transform Data from Wide to Long Format (Melt) ---

# I am using pd.melt() to convert the wide-format final_summary_df into a long-format DataFrame
# suitable for Seaborn grouped bar charts.

# We melt based on the 'Unit' column (id_vars) and stack the rate columns (value_vars).
rate_metrics_df = pd.melt(
    final_summary_df,
    id_vars=['Unit'],
    value_vars=['CF/60_Rate', 'xG/60_Rate', 'GF/60_Rate'], # Using only CF, xG, and GF to match the plot title
    var_name='Metric',
    value_name='Rate_Per_60'
)

# I am cleaning up the 'Metric' names by removing the '_Rate' suffix for cleaner plot labels.
rate_metrics_df['Metric'] = rate_metrics_df['Metric'].str.replace('_Rate', '', regex=False)

print("\nStep 1.5: rate_metrics_df successfully created in Long Format.")
print(rate_metrics_df)
```

```python
# --- Step 2.1: Prepare Shot Map Data (5v4 FILTERED) ---

# I am defining constants for team and player identification.
BRUINS_TEAM = 'BOS' 
PASTRNAK_ID = 8477956

# The raw shots data (df_raw_shots) is already in memory from Step 1.1

# --- 1. Filter Shots for Relevance (The SQL WHERE Clause) ---

# I am filtering for only Bruins shots and only Goal or Shot events, using confirmed column names.
df_shots_filtered = df_raw_shots[
    (df_raw_shots['teamCode'] == BRUINS_TEAM) & 
    (df_raw_shots['event'].isin(['SHOT', 'GOAL', 'MISS'])) 
].copy() 

# --- 2. Filter for 5v4 Strength (THE CRITICAL FIX) ---

# I am defining the function to filter for the 5v4 situation using skater counts.
# This solves the ambiguity of the 'situation' column in the raw data.
def is_5v4_power_play(row):
    # Determine if the shooter's team is home or away
    is_home_shooter = row['teamCode'] == row['homeTeamCode']
    
    if is_home_shooter:
        # Bruins (Home) are on PP if they have 5 skaters and Away has 4
        return (row['homeSkatersOnIce'] == 5) and (row['awaySkatersOnIce'] == 4)
    else:
        # Bruins (Away) are on PP if they have 5 skaters and Home has 4
        return (row['awaySkatersOnIce'] == 5) and (row['homeSkatersOnIce'] == 4)

# I am applying the 5v4 filter to the DataFrame. This is a common operation (like an APPLY in SQL).
df_shots_filtered['is_5v4'] = df_shots_filtered.apply(is_5v4_power_play, axis=1)

# I am now keeping only the events that returned True for my 'is_5v4' function.
df_pp_shots = df_shots_filtered[df_shots_filtered['is_5v4'] == True].copy()


# --- 3. The Coordinate Transformation (Final Fix) ---

# I am defining the necessary function to fix the coordinate system for visualization.
def normalize_coordinates(df):
    
    # Identify shots recorded relative to the OPPOSITE net (X > 100 on the 0-200 scale).
    is_defensive_half = df['xCord'] < 0
    
    # X-COORDINATE FIX (Rink Length)
    df['normalized_xCoord'] = np.where(
        is_defensive_half, df['xCord'] * -1, df['xCord']
    )
    
    # Y-COORDINATE FIX (Rink Width - Side-to-Side Flip)
    df['normalized_yCoord'] = np.where(
        is_defensive_half, df['yCord'] * -1, df['yCord']       
    )
    
    return df

# Apply the transformation to the 5v4 shot data.
df_pp_shots = normalize_coordinates(df_pp_shots)

# --- 4. The Pastrnak ON/OFF Flag (Direct Involvement Proxy) ---

# I am creating the final flag by checking if Pastrnak was the shooter.
df_pp_shots['pastrnak_on_ice'] = np.where(
    df_pp_shots['shooterPlayerId'] == PASTRNAK_ID,
    1, # True (Pastrnak was the shooter)
    0  # False (Someone else shot)
)

# --- 5. Final Shot Map Data Frame ---

# I am selecting the final, clean columns needed for visualization.
shot_map_data = df_pp_shots[[
    'normalized_xCoord', 
    'normalized_yCoord', 
    'pastrnak_on_ice', 
    'event'
]].copy()

# I want to confirm the data is ready for plotting.
print("\nStep 2.1 (Corrected): 5v4 Shot Map Data Preparation Complete.")
print(f"Total 5v4 Shots to Plot: {len(shot_map_data)}")
print(shot_map_data.head())
```

```python
# --- Step 3.1: Visualization (With Annotations) ---

# I am importing the necessary libraries for plotting.
import matplotlib.pyplot as plt
import seaborn as sns
from hockey_rink import NHLRink # Assuming this is used elsewhere

# --- A. Rate Metrics Bar Chart ---

# I am creating the figure and axis object.
plt.figure(figsize=(10, 6))
ax = sns.barplot(x='Metric', y='Rate_Per_60', hue='Unit', data=rate_metrics_df,
                 palette={'Pastrnak On': '#00AEEF', 'Pastrnak Off': '#0D2A2B'})

# --- ANNOTATION (NEW CODE) ---
# I am looping through the containers (the bars) to place the value annotations.
for container in ax.containers:
    # ax.bar_label is the simplest way to add labels in newer Matplotlib versions
    # We format the text to two decimal places
    ax.bar_label(container, fmt='%.2f', fontsize=10, padding=3)

# I'm customizing the labels and saving the chart.
plt.title('PP Efficiency: Shots, xG, and Goals per 60 Minutes',
          fontsize=14, fontweight='bold', color='#0D2A2B')
plt.xlabel('Rate Metric', fontsize=12)
plt.ylabel('Events Per 60 Minutes', fontsize=12)
plt.xticks(rotation=15)

# Legend title updated to reflect On vs Off comparison
plt.legend(title='Pastrnak On vs. Off')

plt.grid(axis='y', linestyle='--', alpha=0.6)

plt.tight_layout()
plt.savefig('Pastrnak_Per60_BarChart_Annotated.png')
```

```python
# --- Step 3.1 (Part B): Shot Map Visualization (PASTRNAK SHOTS, MISSES, GOALS) ---

# 1. Initialize the figure and axis.
fig, ax = plt.subplots(figsize=(10, 8))

# --- CRITICAL FIX: Pass the transparency setting for the 'crease' feature during initialization ---
# We set the 'crease' feature's alpha to 0.3 (30% opaque).
# Using rotation=0 (standard horizontal rink). 
rink = NHLRink(
    rotation=0, 
    # The 'crease' key targets the goal crease, making it transparent in case there are any shots within that area.
    crease={"alpha": 0.3} 
)

# Using display_range="offense" to crop to the zone.
# The draw method now only needs the plotting axes and range.
rink.draw(ax=ax, display_range="offense")

# I am defining the color map for the data using your brand hex codes.
COLOR_MISSED_SHOT = '#0D2A2B'
# ... (rest of the color definitions) ...
COLOR_SHOT_ON_NET = '#00AEEF'
COLOR_GOAL = '#F7941D'
GOAL_EDGE_COLOR = 'red' 


# --- Plotting Pastrnak's Shots, Misses, and Goals (NO CHANGES HERE) ---

# Filter data to ONLY Pastrnak's direct involvement (where he was the shooter)
pastrnak_all_events = df_pp_shots[df_pp_shots['pastrnak_on_ice'] == 1].copy()

# 1. Separate into Misses, Shots on Net, and Goals
pastrnak_misses = pastrnak_all_events[pastrnak_all_events['event'] == 'MISS']
pastrnak_shots_on_net = pastrnak_all_events[pastrnak_all_events['event'] == 'SHOT']
pastrnak_goals = pastrnak_all_events[pastrnak_all_events['event'] == 'GOAL']


# 2. Plot Misses (Lowest layer, smaller size, less opaque)
ax.scatter(pastrnak_misses['normalized_xCoord'], pastrnak_misses['normalized_yCoord'], 
            label=f'Misses ({len(pastrnak_misses)})', 
            alpha=0.4, 
            s=80, 
            color=COLOR_MISSED_SHOT, 
            edgecolors='none', 
            zorder=1) 


# 3. Plot Shots On Net 
ax.scatter(pastrnak_shots_on_net['normalized_xCoord'], pastrnak_shots_on_net['normalized_yCoord'], 
            label=f'Shots on Net ({len(pastrnak_shots_on_net)})', 
            alpha=0.6, 
            s=120, 
            color=COLOR_SHOT_ON_NET, 
            edgecolors='gray', 
            linewidth=0.5,
            zorder=2)


# 4. Plot Goals 
ax.scatter(pastrnak_goals['normalized_xCoord'], pastrnak_goals['normalized_yCoord'], 
            marker='*', 
            s=300, 
            color=COLOR_GOAL, 
            edgecolors=GOAL_EDGE_COLOR, 
            linewidth=1, 
            label=f'Goals ({len(pastrnak_goals)})', 
            zorder=4)

# --- Finalize the plot. ---
ax.set_title(f'{PASTRNAK_NAME} Individual PP Shots & Goals (2024-25)', fontsize=18, fontweight='bold', color='#0D2A2B')

# --- CRITICAL FIX: Move the Legend Outside the Plot ---
ax.legend(
    title='Shot Outcome', 
    loc='upper left',          # Specifies where the anchor point should be located on the legend itself
    bbox_to_anchor=(1.02, 1)  # Defines the coordinates (x, y) on the figure where the legend's anchor point (upper left) should be placed.
)

plt.tight_layout()
plt.savefig('Pastrnak_Individual_ShotMap_Detailed.png')
# plt.show()
```
